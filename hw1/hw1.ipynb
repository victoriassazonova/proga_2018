```python
import random
import re
#выбор пользователем категории
def choice():
    with open('football.txt', 'r', encoding='utf-8') as a, open('cities.txt', 'r', encoding='utf-8') as b, open('oscar.txt', 'r', encoding='utf-8') as c:
     print('Хотите поиграть? Выберите категорию введя "1", "2" или "3". 1) футбольные клубы 2) столицы мира 3) актеры и актриссы, имеющие оскар')
     choice = input()
     while True:
         if choice == '1':
             chosen = a.read()
             break 
         if choice == '2':
             chosen = b.read()
             break 
         if choice == '3':
             chosen = c.read()
             break
         else:
             print('Нужно было ввести номер: 1,2 или 3. Введите номер: ')
             choice = input()
    return chosen            
```

```python
#выбор случайного слова в выбранной категории
def wordchoice(chosen):
    words = chosen.split()
    thewordindex = random.randint(0, len(words) - 1)
    hang = words[thewordindex]
    return hang
```

```python
#вывод количества оставшихся попыток, проверка введенной буквы,замена черточек на буквы, победа или проигрыш
def game(pics, hang):
    wrong = ''
    left=7
    blanks = '_ '*len(hang)
    print(blanks)
    while left >= 1:
        print(pics[len(wrong)])   #сколько пропущенных
        if left == 1:
            print('Осталась ', left, ' попытка')
        if left == 2 or left == 3 or left == 4:
            print('Осталось ', left, ' попытки.')
        if left > 4:
            print('Осталось ', left, ' попыток.') 
        print('Введите букву:')
        guess=input()
        guess = guess.lower()
        hanging=hang.lower()
        if len(guess) != 1:
            print('Введите одну букву!')
            continue
        if guess not in 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя':
            print('Пожалуйста, введите букву кириллицы!')
            continue
        if guess in hanging:
            for i in range(len(hanging)):
                if hanging[i] in guess:
                    blanks = blanks[:i*2] + hanging[i] + ' ' + blanks[(i+1)*2:]
                    print(blanks)
                    u=blanks.find('_')
                    if u == -1:
                        print('Ура, вы угадали')
                        left=-1
        if guess not in hanging:
            wrong = wrong+guess
            left-=1
            print(blanks)
        if left==0:
            print(pics[len(wrong)])
            print('К сожалению вы проиграли, попытки закончились. Это было слово:', hanging)

    print('Игра закончена')  
    want_again()
```

```python
#хочет ли пользователь сыграть еще раз
def want_again():
    print('Еще раз? Введите "да"')
    decision = input()
    if decision.lower() == 'да':
        game(pics, wordchoice(choice()))
    else:
        print('Очень жаль:(')

```

```python
def main():
    game(pics, wordchoice(choice()))
    
if __name__== '__main__':
    main()
```
