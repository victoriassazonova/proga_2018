```python
#!!вставьте где надо ваш токен!! мне так и не удалось выяснить, как я могу сделать это иначе, чтобы выложить

"""users = ('elmiram','maryszmary','lizaku',
    'nevmenandr','ancatmara','roctbb','akutuzov',
    'agricolamz','lehkost','kylepjohnson','mikekestemont',
    'demidovakatya','shwars','JelteF','timgraham',
    'arogozhnikov','jasny','bcongdon','whyisjake')"""
users = ('elmiram','maryszmary','lizaku',
    'nevmenandr','ancatmara','roctbb','akutuzov')

#Выбираем из общего списка одного пользователя, о котором хотим получить информацию
def getUsername(users):
    print('Выберите пользователя:')
    print(users, sep=', ')
    chosen_user = input()
    while True:
        if chosen_user in users:
            print('Отлично, вы выбрали:', chosen_user)
            break
        else:
            print('Такого пользователя нет в списке, введите другое имя: ')
            chosen_user = input('Введите пользователя: ')
    return chosen_user

#Получаем информацию о пользователе
def getData(user):
    token = '0f978cf53defd89cf31c287eed1f769eea6ebab8' 
    page = 1
    url = 'https://api.github.com/users/%s/repos?page=%s&per_page=100&access_token=%s' % (user, page, token)   
    response = urllib.request.urlopen(url)
    text = response.read().decode('utf-8')
    data = json.loads(text)
    while len(data) == 100:
            page += 1
            data += getData(user) 
    return data

#Печатаем репозитории нашего пользователя
def printingRepos(data):
    print('Репозитории пользователя: ')
    print(len(data))
    for i in data:
        print('Название:')
        print(i['name'])
        if (i["description"]) is None:
            print('У этого репозитория нет описания')
        else:
            print('Описание:')
            print(i["description"])
    return data

#Считаем языки пользователя
def getLanguages(data):
    languages = {}
    for i in data: 
        if i['language'] is not None:
            if i['language'] in languages:
                languages[i['language']] += 1
            else:
                languages[i['language']] = 1
    used_languages = []
    for key in languages.keys():
        used_languages.append(key)
    print('Пользователь работает с языками: ', ', '.join(used_languages))
    for key, value in languages.items():
        print('Язык {} используется {} раз'.format(key, value))
    return data
        
#Выясняем, у кого больше всего репозиториев
def mostRepos(users):
    count = {}
    for user in users:
        count[user] = 0
        page = 1
        data = getData(user)
        while len(data) == 100:
            count[user] += len(data)
            page += 1
            data += getData(user) 
        else:
            count[user]+= len(data)
    for key, value in count.items():
        if value == max(count.values()):
            print('Репозиториев больше всего у:', key, '-', value)
    return data

#Выясняем, какой язык самый популярный
def mostPopularLanguage(users):
    languages = {}
    for user in users:
        data = getData(user)
        for i in data:
            if i['language'] is not None:
                if i['language'] in languages:
                    languages[i['language']] += 1
                else:
                    languages[i['language']] = 1
    for key, value in languages.items():
        if value == max(languages.values()):
            print('Самым популярным языком является:', key)

#Выясняем, кто самый "популярный" среди пользователей
def mostPopular(users):
    token = '0f978cf53defd89cf31c287eed1f769eea6ebab8'
    followers={}
    for user in users:
        followers[user] = 0
        page = 1
        url = 'https://api.github.com/users/%s/followers?page=%s&per_page=100&access_token=%s' % (user, page, token)
        response = urllib.request.urlopen(url)
        text = response.read().decode('utf-8')
        data = json.loads(text)
        while len(data) == 100:
            followers[user] += len(data)
            page += 1
            data += getData(user) 
        else:
            followers[user]+= len(data)
    for key, value in followers.items():
        if value == max(followers.values()):
            print('Подписчиков больше всего у:', key, '-', value)

def main():
    getLanguages(printingRepos(getData(getUsername(users))))
    mostRepos(users)
    mostPopularLanguage(users)
    mostPopular(users)

if __name__ == '__main__':
    main()
```
